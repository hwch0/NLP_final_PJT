## 1) 패키지와 모듈 설치 및 input(키워드) 입력

```python
### 참고사이트: https://coding-kindergarten.tistory.com/91

#step1.selenium 패키지와 time 모듈 import
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import numpy as np
import pandas as pd
from selenium.webdriver.chrome.options import Options
options = Options()

# chrome을 전체화면으로 넓히는 옵션입니다.
options.add_argument('--start-fullscreen')


#step2.검색할 키워드 입력 
query = input()

#step3.크롬드라이버로 url로 접속
url = 'https://class101.net/categories/'
driver = webdriver.Chrome(executable_path="/Users/USER/Desktop/datamining/chromedriver.exe", chrome_options=options)  ##크롬드라이버가 설치된 위치
driver.get(url)
time.sleep(3)

#step4.검색창에 키워드 입력 후 엔터
search_box = driver.find_element_by_css_selector(".SearchInput__Input-go705p-2.bxkgtv")
search_box.send_keys(query)
search_box.send_keys(Keys.RETURN)
time.sleep(2)


## 스크롤하기 ##
#ActionChains모듈 가져오기
from selenium.webdriver import ActionChains

#ActionChains생성
action = ActionChains(driver)
```

## 2) 필요한 리스트 만들어주기

```python
## 리스트 리셋할 때 사용

title_list=[] ## 강의명 리스트
intro=[]      ## 강의 소개글을 tag<p> 단위로 받아올 리스트
intro_list=[] ## 강의 소개글을 강의 단위로 저장할 리스트
keyword=[]    ## 강의별 검색 키워드 저장 리스트
```

## 3) 검색 결과의 1페이지에 있는 강의 전체 크롤링 (페이지당 40개)  ---1페이지에 대한 코드만 작성
- 1페이지 당 40개의 강의가 있음
- 상단 8개 / 하단 32개로 섹션이 구분되어 있음
- 상단과 하단 강의를 구분해서 for문을 돌려줌
- ### 페이지를 수동으로 이동 [1] ->[2] ->[3] ...->[5] 하면서 3) ~ 5) 과정을 반복하면 list에 데이터가 누적 저장됩니다.

### 3-1) 1페이지 상단 강의 8개 크롤링

```python
for n in range(1,9):   ## 1페이지 상단 강의1~강의8 크롤링 
    x_path = '//*[@id="wrapper"]/div/div[3]/div/section/div/div[4]/div[3]/div[1]/ul/li[{0}]/div/a'.format(n)
             
    driver.find_element_by_xpath(x_path).click()
    time.sleep(5)
    
        
    intro_text = driver.find_elements_by_tag_name("p") ## 소개글 가져오기
    intro=[]
    
    for t in intro_text:
        text = t.text
        intro.append(text)
    intro_list.append(' '.join(intro)) ## 소개글 가져와서 리스트에 저장
    

    keyword.append(query) ## 키워드 리스트 업데이트
    driver.back() #뒤로가기
    time.sleep(5)   
    
```

### 3-2) 1페이지 하단 강의 32개 크롤링

```python
for n in range(1,33):   ## 1페이지 하단 강의1~강의32까지 크롤링 
    x_path = '//*[@id="wrapper"]/div/div[3]/div/section/div/div[4]/div[3]/div[4]/div/div/ul/li[{0}]/div/a'.format(n)
             
    driver.find_element_by_xpath(x_path).click()
    time.sleep(5)
    
        
    intro_text = driver.find_elements_by_tag_name("p") ## 소개글 가져오기
    intro=[]
    
    for t in intro_text:
        text = t.text
        intro.append(text)
    intro_list.append(' '.join(intro)) ## 소개글 가져와서 리스트에 저장
    

    keyword.append(query) ## 키워드 리스트 업데이트
    driver.back() #뒤로가기
    time.sleep(5)   
    
```

## 4) 1페이지의 강의명 크롤링 (총 40개)

```python
titles = driver.find_elements_by_css_selector(".Card__Title-sc-1esp5o0-4.bVCpyh")

for i in titles:
    title = i.text
    title_list.append(title)
```
