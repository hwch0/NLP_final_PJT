```python
from sklearn.feature_extraction.text import TfidfVectorizer

#2 개 미만의 문서 (min_df) 또는 99 % 이상 (max_df)에 나타나는 용어는 제외합니다.
vec = TfidfVectorizer( max_df=0.90, 
  min_df=2 )
```

```python
features = vec.fit_transform(docs_list)
```

```python
from sklearn.decomposition import NMF
#n_topics = 20
#random_state = 42
nmf = NMF(n_components=20)
nmf.fit(features)
```

```python
enumerate(nmf.components_)
```

```python
def display_top_words(n_top_words, feature_names, nmf):
    for i, topic_vec in enumerate(nmf.components_):
        print(i, end=' ')
    for fid in topic_vec.argsort()[-1:-n_top_words-1:-1]:
        print(feature_names[fid], end=' ')
    print()
```

```python
display_top_words(15, vec.get_feature_names(), nmf)
```

```python
def topic_table(n_top_words, feature_names, nmf):
    topics = {}
    for i, topic_vec in enumerate(nmf.components_):
        topic_descr = ''
        for fid in topic_vec.argsort()[-1:-n_top_words-1:-1]:
            topic_descr = topic_descr + feature_names[fid] + " "
        topics[i] = topic_descr
    return pd.DataFrame({'Top_Topic_Terms': topics})

# Label topics with top 10 terms
topic_df = topic_table(10, vec.get_feature_names(), nmf)

```

## 강의 추천 시스템 함수

```python
def class_recommend(query):
    #크롬드라이버로 url로 접속
    url = 'https://class101.net/categories/'
    driver = webdriver.Chrome(executable_path="/Users/USER/Desktop/datamining/chromedriver.exe")  ##크롬드라이버가 설치된 위치
    driver.get(url)
    time.sleep(3)
    
    #검색창에 키워드 입력 후 엔터
    search_box = driver.find_element_by_css_selector(".SearchInput__Input-go705p-2.bxkgtv")
    search_box.send_keys(query)
    search_box.send_keys(Keys.RETURN)
    time.sleep(2)
    
    # 결과 화면에서 강의명과 url 추출

    titles = driver.find_elements_by_css_selector(".Card__Title-sc-1esp5o0-4.bVCpyh")
    hrefs = driver.find_elements_by_css_selector(".ProductCardfragment__HoverStyledLink-sc-1cja13i-0.bzmYOj")
    titles_list =[]
    hrefs_list =[]
    for i in titles:
        title = i.text
        titles_list.append(title)
    
        for a in hrefs:
            href = a.get_attribute('href')
            if (href not in hrefs_list) :
                hrefs_list.append(href)

    ### 강의명, 링크 데이터프레임으로 만들기            
    data = {"class_title" : titles_list,
           "link" : hrefs_list}
    df = pd.DataFrame(data)
    
    return df
```

```python
## 입력받은 query(예측한 라벨)에 대한 강의를 랜덤으로 3개 가져옴
class_recommend(query).sample(n=3)
```

```python
cmatrix = confusion_matrix(num_target, pred_labels)

#10번째열 까지만 출력
cmatrix[:,:3]
```

```python
# 22개
labels_22 = list(cmatrix.argmax(axis=0))
print(labels_22)
```

```python
#labels_22로 매칭했을 때, 전체 클러스터링 정확도 계산 과정(성능확인 위해서)

i = 0
sum_result = 0
for c in cmatrix.transpose():
    print(c)
    print('label:', labels_22[i])
    print('value:', c[labels_22[i]])
    i += 1
    sum_result += c[labels_22[i]]
    
```
